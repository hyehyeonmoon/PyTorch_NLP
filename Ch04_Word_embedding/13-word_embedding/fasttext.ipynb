```python
import fasttext
```

## FastText model

```python
model=fasttext.train_unsupervised('review.sorted.uniq.refined.tsv.text.tok', model='skipgram', epoch=1, lr=0.1)
print(model.words)
```

```python
print(model['행사']) # get the vector of the word '행사'
print(len(model['행사']))
```

### oov 단어도 출력 가능

```python
model.get_word_vector("보아즈")
```

### Nearest neighbor queries = Measure of similarityPermalink
단어 사이의 유사성을 계산해 최근접 이웃을 찾을 수 있을 수 있습니다. 모든 단어들을 계산해서 가장 유사한 단어 10개를 나타내고 해당 단어가 있다면, 상단에 표시되고 유사도는 1입니다.

```python
model.get_nearest_neighbors('반짝반짝')
```

### Word analogies
세 단어의 관계유추도 가능

```python
model.get_analogies("저렴","싸","할인")
```
